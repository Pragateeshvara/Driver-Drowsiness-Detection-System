```python
import kagglehub
import os
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential , Model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.layers import Conv2D, LSTM, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
```

```python
dataset_path = '/kaggle/input/drowsiness-dataset/train'

# Dictionary to store counts
class_counts = {}

# Count images in each class folder
for class_name in os.listdir(dataset_path):
    class_dir = os.path.join(dataset_path, class_name)
    if os.path.isdir(class_dir):
        count = len(os.listdir(class_dir))
        class_counts[class_name] = count

# Plot
plt.figure(figsize=(8,6))
plt.bar(class_counts.keys(), class_counts.values(), color='skyblue')
plt.title('Number of Images per Class')
plt.xlabel('Class Name')
plt.ylabel('Image Count')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Print exact counts
for cls, count in class_counts.items():
    print(f"{cls}: {count} images")
```

 Load and Preprocess the Data

```python
# Define dataset directory
data_dir = "/kaggle/input/drowsiness-dataset/train"

# Initialize list to store image paths and labels
data = []

# Map each folder name to its label
for label in os.listdir(data_dir):
    folder_path = os.path.join(data_dir, label)
    if os.path.isdir(folder_path):
        for file in os.listdir(folder_path):
            if file.endswith(".jpg"):
                data.append([os.path.join(folder_path, file), label])

df = pd.DataFrame(data, columns=["image_path", "label"])
print(df.head())
print("Class counts:\n", df['label'].value_counts())
```

Train/Val Split and Label Encoding

```python
le = LabelEncoder()
df["encoded_label"] = le.fit_transform(df["label"])
class_names = le.classes_

train_df, val_df = train_test_split(df, test_size=0.2, stratify=df["encoded_label"], random_state=42)
```

ImageDataGenerator with Augmentations

```python
IMG_SIZE = 64
batch_size = 32

# Augmented generator
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    x_col="image_path",
    y_col="label",
    target_size=(IMG_SIZE, IMG_SIZE),
    class_mode="categorical",
    batch_size=batch_size,
    shuffle=True
)

val_gen = val_datagen.flow_from_dataframe(
    dataframe=val_df,
    x_col="image_path",
    y_col="label",
    target_size=(IMG_SIZE, IMG_SIZE),
    class_mode="categorical",
    batch_size=batch_size,
    shuffle=False
)
```

Build CNN Model

```python
cnn_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(4, activation='softmax')
])

cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
cnn_model.summary()
```

Train with Augmented Data

```python
history = cnn_model.fit(
    train_gen,
    epochs=50,
    validation_data=val_gen
)
```

```python
val_gen.reset()
y_pred = cnn_model.predict(val_gen)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = val_gen.classes

print(classification_report(y_true, y_pred_classes, target_names=class_names))
```

```python
cnn_model.save("cnn_drowsiness_model.keras")
```

RNN Model

```python
base_dir = "/kaggle/input/drowsiness-dataset/train"

# Create a dataframe with image paths and labels
data = []
for label in os.listdir(base_dir):
    folder = os.path.join(base_dir, label)
    if os.path.isdir(folder):
        for file in sorted(os.listdir(folder)):  # sorted for consistent sequencing
            if file.endswith(".jpg"):
                data.append([os.path.join(folder, file), label])

df = pd.DataFrame(data, columns=["image_path", "label"])

# Encode labels to 0-3
le = LabelEncoder()
df["encoded_label"] = le.fit_transform(df["label"])
label_map = dict(zip(le.classes_, le.transform(le.classes_)))
print("Label mapping:", label_map)
```

Load Pretrained Feature Extractor (MobileNetV2)

```python
base_model = MobileNetV2(weights='imagenet', include_top=False, pooling='avg')
feature_extractor = Model(inputs=base_model.input, outputs=base_model.output)

def extract_features(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    features = feature_extractor.predict(x, verbose=0)
    return features.flatten()
```

Create Sequences (e.g., 10 images  1 sequence)

```python

sequence_length = 10
step_size = 2  # Create more overlapping sequences

X_seq, y_seq = [], []

for label in df["label"].unique():
    temp = df[df["label"] == label]
    paths = temp["image_path"].values
    encoded_label = temp["encoded_label"].values[0]

    for i in range(0, len(paths) - sequence_length, step_size):
        seq_features = []
        for j in range(sequence_length):
            f_path = paths[i + j]
            features = extract_features(f_path)
            seq_features.append(features)
        X_seq.append(seq_features)
        y_seq.append(encoded_label)

X_seq = np.array(X_seq)
y_seq = np.array(y_seq)
y_seq_cat = to_categorical(y_seq, num_classes=4)

print("Total sequences generated:", len(X_seq))
```

Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_seq, y_seq_cat, test_size=0.2, stratify=y_seq, random_state=42
)
```

Build RNN (LSTM) Model

```python
rnn_model = Sequential([
    LSTM(64, input_shape=(sequence_length, X_seq.shape[2]), return_sequences=False),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dense(4, activation='softmax')
])

rnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
rnn_model.summary()
```

Train RNN

```python
rnn_history = rnn_model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=16,
    validation_split=0.1
)
```

```python
y_pred = np.argmax(rnn_model.predict(X_test, verbose=1), axis=1)
y_true = np.argmax(y_test, axis=1)

print(classification_report(y_true, y_pred, target_names=le.classes_))
```

```python
rnn_model.save("rnn_drowsiness_model.keras")
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(rnn_history.history['accuracy'], label='Train Accuracy')
plt.plot(rnn_history.history['val_accuracy'], label='Val Accuracy')
plt.title('RNN Accuracy over 50 Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(rnn_history.history['loss'], label='Train Loss')
plt.plot(rnn_history.history['val_loss'], label='Val Loss')
plt.title('RNN Loss over 50 Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python

# Pick 5 random sequences from the test set
indices = np.random.choice(len(X_test), size=5, replace=False)

for idx in indices:
    sequence = X_test[idx]  # shape: (10, feature_dim)
    actual_class = np.argmax(y_test[idx])
    predicted_class = np.argmax(rnn_model.predict(np.expand_dims(sequence, axis=0), verbose=0))

    # To display: Show the **middle image** in the 10-frame sequence
    middle_frame_index = sequence_length // 2
    middle_img_path = df[df["encoded_label"] == actual_class]["image_path"].iloc[middle_frame_index]

    # Load & display the image
    img = cv2.imread(middle_img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(img_rgb)
    plt.title(f"Actual: {le.classes_[actual_class]} | Predicted: {le.classes_[predicted_class]}")
    plt.axis("off")
    plt.show()
```

```python
def display_sequence(sequence_paths, actual_class, predicted_class):
    fig, axs = plt.subplots(1, len(sequence_paths), figsize=(20, 3))
    for i, path in enumerate(sequence_paths):
        img = cv2.imread(path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        axs[i].imshow(img)
        axs[i].axis('off')
        axs[i].set_title(f'Frame {i+1}')
    plt.suptitle(f"Actual: {le.classes_[actual_class]} | Predicted: {le.classes_[predicted_class]}")
    plt.tight_layout()
    plt.show()
```
